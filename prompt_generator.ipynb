```python
import openai
from openai import OpenAI
import pickle

print(openai.VERSION)
```

```python
client = OpenAI(api_key='')
```

```python
prompts =[]
```

```python
for i in range(300):
  response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
      {"role": "system", "content": "You are a helpful prompt generator. You help the user by generating prompts for them to use with LLMs."},
      {"role": "user", "content": "I want a prompts describing a website on a random theme. Respond only with the prompt."}
    ]
  )
  prompts.append(response.choices[0].message.content.strip().split("\n")[0])
```

```python
txt_file = open("prompts.txt", "a")
for e in prompts:
    txt_file.write(e + '\n')
txt_file.close()
```
